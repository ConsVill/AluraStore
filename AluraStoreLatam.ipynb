```python
from matplotlib import pyplot as plt


ventas_por_categoria_1 = tienda["Categoría del Producto"].value_counts()
ventas_por_categoria_2 = tienda2["Categoría del Producto"].value_counts()
ventas_por_categoria_3 = tienda3["Categoría del Producto"].value_counts()
ventas_por_categoria_4 = tienda4["Categoría del Producto"].value_counts()

categorias= ["Muebles", "Electrónicos", "Juguetes", "Deportes y diversión", "Electrodomésticos", "Artículos para el hogar", "Libros", "Instrumentos musicales"]

ventas1 = [ventas_por_categoria_1.get(cat, 0) for cat in categorias]
ventas2 = [ventas_por_categoria_2.get(cat, 0) for cat in categorias]
ventas3 = [ventas_por_categoria_3.get(cat, 0) for cat in categorias]
ventas4 = [ventas_por_categoria_4.get(cat, 0) for cat in categorias]

resultados = {
    "tienda 1": ventas1,
    "Tienda 2": ventas2,
    "Tienda 3": ventas3,
    "Tienda 4": ventas4
}

import matplotlib.pyplot as plt
import numpy as np


# Función de gráfico
def survey(results, category_names):
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.colormaps['tab20'](
        np.linspace(0.0, 1.0, data.shape[1]))

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.5,
                        label=colname, color=color)

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'black'
        ax.bar_label(rects, label_type='center', color=text_color)
    ax.legend(ncols=2, bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')

    return fig, ax

# Llamada a la función para mostrar el gráfico
survey(resultados, categorias)
plt.show()
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

df = pd.concat([tienda], ignore_index=True)

df["Fecha de Compra"] = pd.to_datetime(df["Fecha de Compra"], dayfirst=True) # Asegura tipo datetime

ventas_anuales = df.groupby(df['Fecha de Compra'].dt.to_period('Y')) ['Precio'].mean().reset_index()
ventas_anuales['Fecha de Compra'] = ventas_anuales['Fecha de Compra'].dt.to_timestamp()

print(f"La cantidad promedio vendida por año es: {ventas_anuales}")
###

#Tienda 2

df = pd.concat([tienda2], ignore_index=True)

df["Fecha de Compra"] = pd.to_datetime(df["Fecha de Compra"], dayfirst=True) # Asegura tipo datetime

ventas_anuales2 = df.groupby(df['Fecha de Compra'].dt.to_period('Y')) ['Precio'].mean().reset_index()
ventas_anuales2['Fecha de Compra'] = ventas_anuales2['Fecha de Compra'].dt.to_timestamp()

print(f"La cantidad promedio vendida por año en la tienda 2 es: {ventas_anuales2}")
###

#Tienda 3

df = pd.concat([tienda3], ignore_index=True)

df["Fecha de Compra"] = pd.to_datetime(df["Fecha de Compra"], dayfirst=True) # Asegura tipo datetime

ventas_anuales3 = df.groupby(df['Fecha de Compra'].dt.to_period('Y')) ['Precio'].mean().reset_index()
ventas_anuales3['Fecha de Compra'] = ventas_anuales3['Fecha de Compra'].dt.to_timestamp()

print(f"La cantidad promedio vendida por año en tienda 3 es: {ventas_anuales3}")
###

#Tienda 4

df = pd.concat([tienda], ignore_index=True)

df["Fecha de Compra"] = pd.to_datetime(df["Fecha de Compra"], dayfirst=True) # Asegura tipo datetime

ventas_anuales4 = df.groupby(df['Fecha de Compra'].dt.to_period('Y')) ['Precio'].mean().reset_index()
ventas_anuales4['Fecha de Compra'] = ventas_anuales4['Fecha de Compra'].dt.to_timestamp()

print(f"La cantidad promedio vendida por año en lat ienda 4 es: {ventas_anuales4}")


#Grafico

import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde las URLs
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer CSVs y agregar etiqueta de tienda
tienda1 = pd.read_csv(url1)
tienda1['Tienda'] = 'Tienda 1'

tienda2 = pd.read_csv(url2)
tienda2['Tienda'] = 'Tienda 2'

tienda3 = pd.read_csv(url3)
tienda3['Tienda'] = 'Tienda 3'

tienda4 = pd.read_csv(url4)
tienda4['Tienda'] = 'Tienda 4'

# Unir todos los datos en un solo DataFrame
df = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

# Asegurar tipo datetime
df['Fecha de Compra'] = pd.to_datetime(df['Fecha de Compra'], dayfirst=True)

# Agrupar por año y tienda, y calcular el promedio de ventas
df['Año'] = df['Fecha de Compra'].dt.year
ventas_anuales = df.groupby(['Año', 'Tienda'])['Precio'].mean().reset_index()

# Pivotear para formato gráfico
pivot = ventas_anuales.pivot(index='Año', columns='Tienda', values='Precio')

# Graficar
plt.figure(figsize=(10, 6))
for tienda in pivot.columns:
    plt.plot(pivot.index, pivot[tienda], marker='o', label=tienda)

plt.title('Promedio de ventas anuales por tienda')
plt.xlabel('Año')
plt.ylabel('Promedio de Precio')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
